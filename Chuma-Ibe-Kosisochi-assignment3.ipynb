```python
#data.groupby("Congress Dist")["Sqft_per_floor"].describe()
stats = data.groupby("Congress Dist")["Sqft_per_floor"].describe()
plt.rcParams["figure.dpi"] = 150
plt.plot(stats["mean"], marker='+', linewidth = 1.0, label = "Average sqft per floor")
#plt.xlim(0,19)
plt.xticks(range(0, 20, 1))
plt.xlabel("Congressional District")
plt.ylabel("Square Footage per floor")
plt.legend()
plt.title('Average Square Footage per floor by Congressional District')
####################################################
###  file containing writeup uploaded separately ### 
####################################################
```
