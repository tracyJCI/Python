```python
x_sc = bqplot.LinearScale()
y_sc = bqplot.OrdinalScale()
c_sc = bqplot.ColorScale(scheme = "BuPu")
xa = bqplot.Axis(scale = x_sc)
ya = bqplot.Axis(scale = y_sc, orientation = 'vertical')
ca = bqplot.ColorAxis(scale = c_sc, orientation = 'vertical', side = 'right')

mmap = bqplot.market_map.MarketMap(color = dogN.values, names = dogN.index.values,
                                   scales = {'color': c_sc}, axes = [ca])

myLabel = ipywidgets.Label()

def get_label(change):
    i = change['new'][-1:][0]
    v = dogN[i]
    myLabel.value = "Total count = " + str(v)
    
mmap.observe(get_label, 'selected')

ipywidgets.VBox([myLabel, ipywidgets.HBox([mmap])])
```

```python
x_sc2 = bqplot.LinearScale(min = 1998, max = 2017)
y_sc2 = bqplot.LinearScale()
xa2 = bqplot.Axis(scale = x_sc2, label = "Year")
ya2 = bqplot.Axis(scale = y_sc2, label = "Total counts", orientation = 'vertical')

i = "BELLA"
sel = (df["AnimalName"] == i)
aa = df[sel]
aa = aa.groupby("year")["RowNumber"].count()

right = bqplot.Lines(x = aa.index, y = aa.values, scales = {'x': x_sc2, 'y': y_sc2}, colors = ['#B9D0F2'])
fig2 = bqplot.Figure(title = "by Year", axes = [xa2, ya2], marks = [right])


myLabel = ipywidgets.Label()

def get_label(change):
    i = change['new'][-1:][0]
    v = dogN[i]
    myLabel.value = "Total count = " + str(v)
    
mmap.observe(get_label, 'selected')

def showYear(change):
    i = change['new'][-1:][0]
    sel = (df["AnimalName"] == i)
    aa = df[sel]
    aa = aa.groupby("year")["RowNumber"].count()
    #bb = aa.values
    #cc = aa.index
    right.x = aa.index
    right.y = aa.values
    #print(change)
    
mmap.observe(showYear, 'selected')    
ipywidgets.VBox([myLabel, ipywidgets.HBox([mmap, fig2])])
```
