```python
stats = data.groupby("Year Acquired")["Year Constructed"].describe()
plt.rcParams["figure.dpi"] = 150
plt.plot(stats["max"], marker='+', linewidth = 1.0, label = "Max")
plt.plot(stats["min"], marker='x', linewidth = 1.0, label = "Min")
plt.fill_between(stats.index, stats["min"], stats["max"], color = "#dddddd")
plt.xlabel("Year Acquired")
plt.ylabel("Year Constructed")
plt.legend()
plt.title('Maximum and Minimum Constructed Year by Acquired Year')
####################################################
###  file containing writeup uploaded separately ### 
####################################################
```

```python
plt.rcParams["figure.dpi"] = 150
stats2 = data.groupby("Congress Dist")["Square Footage"].sum()
plt.plot(stats2, marker='+', linewidth = 1.0, label = "Total Square Footage")
plt.xlim(0,19)
plt.yscale("log")
plt.xlabel("Congressional District")
plt.ylabel("Square Footage")
plt.legend()
plt.title('Total Square Footage by Congressional District')
####################################################
###  file containing writeup uploaded separately ### 
####################################################
```
