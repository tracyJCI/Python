```python
#MARKET MAP

#scales
x_sc, y_sc = bqplot.OrdinalScale(), bqplot.OrdinalScale() # note, just a different way to call things
c_sc = bqplot.ColorScale(scheme="Blues")

#axes
c_ax = bqplot.ColorAxis(scale = c_sc, orientation = 'vertical')


import matplotlib.cm as cm
cmap = cm.Blues

mmap = bqplot.market_map.MarketMap(color = br.values, 
                                  names = br.index.values,
                                  scales={'color':c_sc},
                                   axes=[c_ax],
                                   title = 'NYC Boroughs by Dogs')
mySelectedLabel = ipywidgets.Label()

def get_data_value(change):
    i = change['new'][-1:][0]
    v = br[i]
    mySelectedLabel.value = 'Borough = ' + str(i) + ', Number of Dogs =' + str(v) 
    
mmap.observe(get_data_value, 'selected')


#BAR GRAPH

#scales
x_scl = bqplot.OrdinalScale()
y_scl = bqplot.LinearScale()

#axes
x_axis = bqplot.Axis(scale=x_scl, label='Breed', tick_rotate = 10, tick_style = {'font-size': 10}
                    )
y_axis = bqplot.Axis(scale=y_scl, orientation='vertical', label='Number of Dogs')

axes_options = {'x': {'label': 'X'}, 'y': {'label': 'Y'}}

i = 'Manhattan'
mask = (dogs['Borough'] == i)

dogbreeds = dogs[mask]
dogbreeds = dogbreeds.groupby('BreedName')['RowNumber'].count()
dogbreeds = dogbreeds.sort_values()
dogbreeds = dogbreeds.nlargest(n=5)
breed_number = dogbreeds.values
breeds = dogbreeds.index

bars = bqplot.Bars(x=breeds, y=breed_number, scales={'x': x_scl, 'y': y_scl}, axes_options=axes_options)


fig_bar = bqplot.Figure(marks = [bars], axes= [x_axis, y_axis], 
                        title= 'Most Popular Dog Breeds')


def get_data_value2(change):
    i = change['new'][-1:][0]
    mask = (dogs['Borough'] == i)
    dogbreeds = dogs[mask]
    dogbreeds = dogbreeds.groupby('BreedName')['RowNumber'].count()
    dogbreeds = dogbreeds.sort_values()
    dogbreeds = dogbreeds.nlargest(n=5)
    breed_number = dogbreeds.values
    breeds = dogbreeds.index
    bars.x=breeds
    bars.y=breed_number
    

mmap.observe(get_data_value2, 'selected')


ipywidgets.VBox([mySelectedLabel, ipywidgets.HBox([mmap,fig_bar])])
```
